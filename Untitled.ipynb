```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('OnlineRetail.csv' , encoding = 'latin1')

data.head(10)
```

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])

data['InvoiveYearMonth'] = data['InvoiceDate'].map(lambda date: 100 * date.year + date.month)

data.head(10)
```

```python
data['Revenue'] = data['UnitPrice'] * data['Quantity']

data.head(10)
```

```python
month_revenue = data.groupby(['InvoiveYearMonth'])['Revenue'].sum().reset_index()

month_revenue
```

```python
plt.figure(figsize = (15,5))
plt.plot(month_revenue['InvoiveYearMonth'].to_numpy(dtype=str) , month_revenue['Revenue'] , marker = 'o')
plt.grid()
plt.title('Montly Revenue')
```

```python
month_revenue['MonthlyGrowth'] = month_revenue['Revenue'].pct_change()

month_revenue

```

```python
plt.figure(figsize = (15,5))
x =  month_revenue.query('InvoiveYearMonth < 201112')['InvoiveYearMonth'].to_numpy(dtype=str)
y =  month_revenue.query('InvoiveYearMonth < 201112')['MonthlyGrowth'] 
plt.plot(x,y)
plt.grid()
```

```python
country = data.query('Country == "United Kingdom"').reset_index(drop=True)

monthly_active  = country.groupby('InvoiveYearMonth')['CustomerID'].nunique().reset_index()

monthly_active.head(10)
```

```python
plt.figure(figsize = (15,5))
plt.bar(month_revenue['InvoiveYearMonth'].to_numpy(dtype=str) , monthly_active['CustomerID'])
plt.grid(axis = 'y')
```

```python
monthly_sales = country.groupby('InvoiveYearMonth')['Quantity'].sum().reset_index()

monthly_sales
```

```python
plt.figure(figsize = (15,5))
plt.bar(month_revenue['InvoiveYearMonth'].to_numpy(dtype=str) , monthly_sales['Quantity'])
plt.grid(axis = 'y')
```
